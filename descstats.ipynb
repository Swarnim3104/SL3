```python
df = pd.read_csv("C:/Users/Admin/Downloads/archive (2)/employee_data.csv")
df
```

```python
df.head()
```

```python
def categorize(age):
    if 20<=age<30:
        return '20-30'
    if 30<=age<40:
        return '30-40'
    if 40<=age<50:
        return '40-50'
    if 50<=age<60:
        return '50-60'
    else:
        return 'Unknown'
```

```python
df['AgeGroup'] = df['age'].apply(categorize)
```

```python
df
```

```python
statistics = df.groupby('AgeGroup')['salary'].describe()
```
